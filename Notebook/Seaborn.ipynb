# **Seaborn**

# Importing Seaborn

```python
import seaborn as sns
```

Data fiksi dapat digenerate dari: http://roycekimmons.com/tools/generated_data

```python
import numpy as np
import pandas as pd
df = pd.read_csv("dm_office_sales.csv")
df
```

# Scatter Plots
Scatter plot dapat digunakan untuk memperlihatkan bagaimana relasi satu sama lain antara fitur/variable pada data.

```python
sns.scatterplot(x='salary',y='sales',data=df);
```

Seperti yang telah dijelaskan sebelumnya, seaborn sendiri dibuat berdasarkan Matplotlib, sehingga kita bisa menggunakan fungsi-fungsi pada matplotlib untuk mengatur figure yang dipanggil dari seaborn

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df)
```

## Hue
`hue` dapat digunakan untuk memberikan warna untuk membedakan fitur kategorikal

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,hue='division')
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience')
```

## Palette
`palette` dapat didefine untuk menentukan palet warna seperti apa yang diinginkan, seperti `cmap` yang telah dijelaskan sebelumnya pada matplotlib, https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,hue='work experience', palette='winter')
```

## Size
`size` bisa digunakan jika kita besar pointnya didasari pada suatu variabel tertentu

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,size='work experience')
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,s=150)
```

## Linewidth and Transparency

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,s=150,linewidth=0.3,alpha=0.4)
```

## Point Style

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education')
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=70)
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(x='salary',y='sales',data=df,style='level of education',hue='level of education',s=100)

# Call savefig in the same cell
plt.savefig('example_scatter.png')
```

# Distribution Plot
Ada banyak sekali metode untuk menamplikan distribusi/persebaran suatu fitur pada data, tetapi kita akan melihat 3 plot yang cukup esensial untuk diketahui

## Rugplot
Rugplot adalah plot dengan bentuk paling sederhana untuk melihat persebaran nilai data, karena hanya membutuhkan satu array data yang akan diplot tersebar berdasarkan nilainya

Note: sumbu y tidak menyatakan/memberikan informasi apapun

Jenis plot ini cukup nice to know tapi jarang digunakan

```python
sns.rugplot(x='salary',data=df)
```

```python
sns.rugplot(x='salary',data=df,height=0.8, label='salary')
sns.rugplot(x='sales',data=df,height=0.4,label='sales')
plt.xlabel('value')
plt.legend()
```

## Distribution Plot dan Histogram Plot
Rugplot akan kurang informatif untuk jumlah data yang besar, karena banyak sekali garis yang bersinggungan yang sulit untuk dilihat.

Sehingga, model plot persebaran data seperti `displot()` dan `histplot()` akan lebih berguna karena akan menunjukkan persebaran count value dari nilai tertentu layaknya histogram

parameter `kde` merupakan KDE (Kernel Density Estimation) https://en.wikipedia.org/wiki/Kernel_density_estimation yang merupakan estimasi probability density function dari sebuah variabel random

###`displot()`

```python
sns.displot(data=df,x='salary',kde=True)
```

```python
sns.displot(data=df,x='salary')
```

### `histplot()`

```python
sns.histplot(data=df,x='salary')
```

```python
sns.histplot(data=df,x='salary',bins=10)
```

```python
sns.histplot(data=df,x='salary',bins=100)
```

## Styling
https://seaborn.pydata.org/generated/seaborn.set_theme.html#seaborn.set_theme

### Style dan Theme pada Seaborn

```python
sns.set(style='darkgrid')
sns.histplot(data=df,x='salary',bins=100)
```

```python
sns.set(style='white')
sns.histplot(data=df,x='salary',bins=100)
```

Untuk mengembalikan style ke default

```python
sns.reset_defaults()
```

### Parameter yang telah dibahas di Matplotlib bisa ditambahkan

```python
sns.displot(data=df,x='salary',bins=20,kde=False,
            color='green',edgecolor='black',lw=5,ls='--')
```

## KDE Plot

```python
sns.kdeplot(data=df['work experience'])
```

```python
    sns.kdeplot(data=df['work experience'], bw_adjust=0.3,color='r')
    sns.kdeplot(data=df['work experience'],bw_adjust=0.5,color='g')
    sns.kdeplot(data=df['work experience'],bw_adjust=0.9,color='b')
```

```python
sns.kdeplot(data=df['work experience'],bw_adjust=0.5,color='r',shade=True,linewidth=3)
```

## 2D KDE Plot

```python
sns.kdeplot(data=df,x='work experience',y='salary', bw_adjust=0.8)
```

# Categorical Plots
Kemudian, bagaimana dengan data 
kategorikal? apakah bisa direpresentasikan dalam plot?

## Countplot
Plot biasa yang menunjukkan count dari setiap kategori

```python
plt.figure(figsize=(10,5))
sns.countplot(x='division',data=df)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='level of education',data=df)
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='division',data=df, hue='level of education')
```

```python
plt.figure(figsize=(10,4))
sns.countplot(x='division',data=df, hue='level of education', palette='Accent')
```

## Barplot
Barplot mirip seperti fungsi groupby yang telah dikenalkan pada tutorial pandas, kita bisa memvisualisasikan kategori data dengan suatu fungsi tertentu seperti mean/rata-rata dan lainnya

Garis hitam yang ada pada barplot merepresentasikan ketidak pastian persebaran datanya yaitu standar deviasinya atau bisa juga di set dari parameter `sd`

```python
plt.figure(figsize=(10,4))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd')
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='level of education',y='salary',data=df,estimator=np.mean,ci='sd',hue='division')
```

Violin plot mirip seperti box plot, hanya saja KDE nya diperlihatkan untuk menggantikan box and whiskers pada box plot

# Comparison Plots
Ada plot yang khusus untuk membandingkan persebaran nilai antar variabel seperti `jointplot()` dan `pairplot()`

## Joint plot

```python
sns.jointplot(x='writing score',y='reading score',data=df)
```

```python
sns.jointplot(x='writing score',y='reading score',data=df,kind='kde')
```

```python
sns.jointplot(x='writing score',y='reading score',data=df,kind='hex')
```

## Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='gender',palette='spring')
```

```python
sns.pairplot(df,hue='gender',palette='bwr_r',corner=True)
```

```python
sns.pairplot(df,hue='gender',palette='bwr_r',kind='kde')
```

# Grid of Plots
Kita bisa mengkombinasikan berbagai jenis plot menjadi suatu grid yang berisi plot-plot

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot,color='g')
g = g.map_diag(sns.kdeplot, lw=2)
g = g.map_lower(sns.kdeplot, colors='r')
```
